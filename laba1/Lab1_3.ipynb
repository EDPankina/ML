3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def norm_matrix(matrix):
    norm_matrix = matrix / np.sum(matrix)
    return norm_matrix

X = np.random.randint(1, 21, size=(3, 3))
print(X,'\n')
norm_X = norm_matrix(X)
print(norm_X)

```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def pyramid_volume(x, y, z):
    return (np.abs(np.dot(x, np.cross(y, z)))) / 6

x = np.array([1, 3, 2])
y = np.array([7, 5, 9])
z = np.array([2, 0, 6])

vol = pyramid_volume(x, y, z)
print('Объем пирамиды равен', vol)
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def interesting_matrix(N):
    im = np.zeros((N, N))
    
    im[0, :] = 1
    im[-1, :] = 1
    im[:, 0] = 1
    im[:, -1] = 1
    
    return im

N = 6
im = interesting_matrix(N)
print(im)
```

7. Как найти наиболее близкое значение (к заданному скаляру) в векторе?

```python
def near_value(vec, scalar):
    ind = np.argmin(np.abs(vec - scalar))
    return vec[ind]

vec = np.array([2.8, 3.5, 1.9, 7.1, 9.0])
scalar = 6.0

nearest_value = near_value(vec, scalar)
print("Наиболее близкое значение:", nearest_value)
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
def CMatrix(X, Y):
   C = 1 / (X[:, None] - Y[None, :])
   return C

X = np.random.randint(100, size=4)
Y = np.random.randint(100, size=4)

print(CMatrix(X, Y))
```

6. Сравните распределения кол-ва смертей на 1000 человек в Европе и Африке

```python
import seaborn as sns

def calc_deaths(row):
  return (row['total_deaths'] / row['population']) * 1000

dataset['deaths_per_million'] = dataset.apply(calc_deaths, axis=1)
dataset['deaths_per_million'] = dataset['deaths_per_million'].fillna(dataset['deaths_per_million'].mean())

europe = dataset[dataset['continent'] == 'Europe']['deaths_per_million']
africa = dataset[dataset['continent'] == 'Africa']['deaths_per_million']

plt.figure(figsize=(10, 6))
sns.histplot(europe, bins=5, color='lightblue', kde=True, label='Европа', stat='density', alpha=0.5)
sns.histplot(africa, bins=5, color='salmon', kde=True, label='Африка', stat='density', alpha=0.5)
plt.title('Сравнение распределения смертей на 1000 человек в Европе и Африке')
plt.xlabel('Смерти на 1000 человек')
plt.ylabel('Плотность')
plt.legend()
plt.grid()

plt.show()
```
